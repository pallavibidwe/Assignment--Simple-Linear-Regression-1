# 1) Delivery_time -> Predict delivery time using sorting time .
Build a simple linear regression model by performing EDA and do necessary transformations and select the best model using R or Python.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.formula.api as smf
```

```python
delivery=pd.read_csv('D:\Data Science\Linear Regresion\Assignment//delivery_time.csv')
delivery
```

# EDA and Data Visualization

```python
delivery.info()
```

```python
sns.distplot(delivery['Delivery Time'])
```

```python
sns.distplot(delivery['Sorting Time'])
```

# Feature Engineering

```python
# Renaming Columns
delivery=delivery.rename({'Delivery Time':'delivery_time', 'Sorting Time':'sorting_time'},axis=1)
delivery
```

# Correlation Analysis

```python
delivery.corr()
```

```python
sns.regplot(x=delivery['sorting_time'],y=delivery['delivery_time'])
```

# Model Building

```python
model=smf.ols("delivery_time~sorting_time",data=delivery).fit()
```

# Model Testing

```python
# Finding Coefficient parameters
model.params
```

```python
# Finding tvalues and pvalues
model.tvalues , model.pvalues
```

```python
# Finding Rsquared Values
model.rsquared , model.rsquared_adj
```

# Model Predictions

```python
# Manual prediction for say sorting time 5
delivery_time = (6.582734) + (1.649020)*(5)
delivery_time
```

```python
# Automatic Prediction for say sorting time 5, 8
new_data=pd.Series([5,8])
new_data
```

```python
data_pred=pd.DataFrame(new_data,columns=['sorting_time'])
data_pred
```

```python
model.predict(data_pred)
```

# 2) Salary_hike -> Build a prediction model for Salary_hike

```python
salary=pd.read_csv('D:\Data Science\Linear Regresion\Assignment//Salary_Data.csv')
salary
```

```python
salary.info()
```

# EDA and Data Visualization

```python
sns.distplot(salary['YearsExperience'])
```

```python
sns.distplot(salary['Salary'])
```

# Correlation Analysis

```python
salary.corr()
```

```python
sns.regplot(x=salary['YearsExperience'],y=salary['Salary'])
```

# Model Building

```python
model=smf.ols("Salary~YearsExperience",data=salary).fit()
```

# Model Testing

```python
model.params
```

```python
# Finding Pvalues and tvalues
model.tvalues, model.pvalues
```

```python
# Finding Rsquared values
model.rsquared , model.rsquared_adj
```

# Model Predictions

```python
# Manual prediction for say 3 Years Experience
Salary = (25792.200199) + (9449.962321)*(3)
Salary
```

```python
# Automatic Prediction for say 3 & 5 Years Experience 
new_data=pd.Series([3,5])
new_data
```

```python
data_pred=pd.DataFrame(new_data,columns=['YearsExperience'])
data_pred
```

```python
model.predict(data_pred)
```
